pandas es un paquete de Python que proporciona estructuras de datos similares a los dataframes de R. Pandas depende de Numpy, la librería que añade un potente tipo matricial a Python. Los principales tipos de datos que pueden representarse con pandas son:

Datos tabulares con columnas de tipo heterogéneo con etiquetas en columnas y filas.

Series temporales.

Pandas proporciona herramientas que permiten:

leer y escribir datos en diferentes formatos: CSV, Microsoft Excel, bases SQL y formato HDF5

seleccionar y filtrar de manera sencilla tablas de datos en función de posición, valor o etiquetas

fusionar y unir datos

transformar datos aplicando funciones tanto en global como por ventanas

manipulación de series temporales

hacer gráficas

En pandas existen tres tipos básicos de objetos todos ellos basados a su vez en Numpy:

Series (listas, 1D),

DataFrame (tablas, 2D) y

Panels (tablas 3D).

## Ejercicio 1
Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla una serie con los datos de las ventas indexada por los años, antes y después de aplicarles un descuento del 10%.

```python
#Vamos de nuevo con este ejercicio, lo haremos muy rapido :D

print("Ventas")
inicio = int(input("¿Cuàl es el primer año "))
fin = int(input("¿Cuàl es el ùltimo año "))

#Ahora vamos a meter los datos en el diccionario
dicc = {}

for i in range(inicio, fin+1):
    dicc[i] = float(input("Ingresa la venta del año"))

#Ahora hacemos el cambio de diccionario a serie de numpy
Serie = pd.Series(dicc)
Serie_desc = Serie*0.9

print("Las ventas son las siguientes")
print(Serie)

print("Las ventas con descuento son las siguientes")
print(Serie_desc)
```

# Ejercicio 3
Escribir una función que reciba una diccionario con las notas de los alumnos en curso en un examen y devuelva una serie con las notas de los alumnos aprobados ordenadas de mayor a menor.

```python
print("Ingresa las calificaciones")
#Definimos primero el diccionario 
dicc = {}

num = int(input("Ingresa el nùmero de alumnos "))

for i in range(num):
    dicc[i]=float(input("Ingresa la calicacion de cada uno "))

#Hacemos el cambio de diccionario a serie de pandas
SerieCal = pd.Series(dicc)
Orden = SerieCal.sort_values(ascending=False)

print("Las calificaciones ordenadas de mayor a menor son")
print(Orden[Orden > 5])
```

# Ejercicio 5
Escribir una función que reciba un DataFrame con el formato del ejercicio anterior, una lista de meses, y devuelva el balance (ventas - gastos) total en los meses indicados.

```python
dicc = {"Mes": ["Enero", "Marzo", "Abril", "Mayo"], "Ventas": [30500, 35600, 28300, 33900], "Gastos":[22000, 23400, 18100, 20700]}
df= pd.DataFrame(dicc)

#Esta es una transformaciòn de datos con calculos 
#Se define la funciòn 

def balance(fila):
    resultado = fila["Ventas"]-fila["Gastos"]
    return resultado 
df["Balance"] = df.apply(balance, axis = 1)

print(df)
```

```python
# Construir una función que construya un DataFrame a partir del un fichero con el formato anterior y devuelva otro DataFrame con el mínimo, el máximo y la media de dada columna.

def ejercicio6(data):
    datos = pd.read_csv(data, sep = ";", decimal=',', thousands='.', index_col=0)
    df = pd.DataFrame(datos)
    resultados = pd.DataFrame([df.min(), df.max(), df.mean()], index=['Mínimo', 'Máximo', 'Media'])
    return resultados
```
